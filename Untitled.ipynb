```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
df = pd.read_excel("stats.xlsx")
```

```python
plt.figure(figsize=(16,8))
i=5
car = df[df.CarId==i]
carX=np.arange(len(car))
carY1=car.AverageSpeed
carY2=car.NumberPoints
carY3=car.Accidents
tit="Car "+str(i+1)+" speed and points for each leap"
plt.title(tit, fontsize=30)
plt.xlabel("Leap", fontsize=15)
plt.plot(carX, carY1, label='Speed', color='b')
plt.plot(carX, carY2/10, label='Points / 10', color='g')
plt.plot(carX, carY3, label='Accidents', color='brown')
plt.plot(carX, [50 for i in carX], label='Limit', color='r')
plt.legend()
```
