```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import colors as mcolors
%matplotlib inline
plt.style.use('ggplot')
df = pd.read_excel("stats90.xlsx")
```

```python
90*250/80
```

```python
ratio=90/250
```

```python
#SPEED

plt.figure(figsize=(16,8))
nbCars=24
colors = dict(mcolors.BASE_COLORS, **mcolors.CSS4_COLORS)
col = list(colors.keys())[8:]
for i in range(nbCars):
    car = df[df.CarId==i]
    carX=np.arange(len(car))
    carY=car.AverageSpeed
    tit="CAR SPEED"
    plt.title(tit, fontsize=30)
    plt.xlabel("Leap", fontsize=15)
    lab="Car"+str(i+1)
    plt.plot(carX, carY*ratio, label=lab, color=col[i+5])
    

plt.plot(np.arange(23), [80 for i in range(23)], label='Limit')
plt.legend()


#POINTS

plt.figure(figsize=(16,8))

for i in range(nbCars):
    car = df[df.CarId==i]
    carX=np.arange(len(car))
    carY=car.NumberPoints
    tit="CAR POINTS"
    plt.title(tit, fontsize=30)
    plt.xlabel("Leap", fontsize=15)
    lab="Car"+str(i+1)
    plt.plot(carX, carY, label=lab, color=col[i+5])
    
plt.legend()


#ACCIDENTS

plt.figure(figsize=(16,8))
for i in range(nbCars):
    car = df[df.CarId==i]
    carX=np.arange(len(car))
    carY=car.Accidents
    tit="CAR ACCIDENTS"
    plt.title(tit, fontsize=30)
    plt.xlabel("Leap", fontsize=15)
    lab="Car"+str(i+1)
    plt.plot(carX, carY, label=lab, color=col[i+5])
    
plt.legend()


#CONFIDENCE

plt.figure(figsize=(16,8))
for i in range(nbCars):
    car = df[df.CarId==i]
    carX=np.arange(len(car))
    carY=car.Confidence
    tit="CAR CONFIDENCE"
    plt.title(tit, fontsize=30)
    plt.xlabel("Leap", fontsize=15)
    lab="Car"+str(i+1)
    plt.plot(carX, carY, label=lab, color=col[i+5])
    
plt.legend()
```

```python
plt.figure(figsize=(16,8))
total=[]
for i in range(22):
    car = df[df.Leap==i]
    total.append(car.NumberPoints.sum())

plt.title("Evolution of the total number of points per leap", fontsize=25)
plt.xlabel("Leap", fontsize=15)
plt.ylabel("Number of points", fontsize=15)
plt.plot(np.arange(22), total)
```

```python
plt.figure(figsize=(16,8))
total=[]
for i in range(1,23):
    car = df[df.Leap==i]
    total.append(car.AverageSpeed.mean()*ratio)

plt.title("Evolution of the average speed per leap", fontsize=25)
plt.xlabel("Leap", fontsize=15)
plt.ylabel("Number of points", fontsize=15)
plt.plot(np.arange(22), total)
```
